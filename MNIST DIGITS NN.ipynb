```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import BatchNormalization,Dropout
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
(xtrain,ytrain),(xtest,ytest) = keras.datasets.mnist.load_data()
```

```python
xtrain = xtrain.astype("float32")/255
xtest = xtest.astype("float32")/255
```

```python
xtrainFlattened = xtrain.reshape(60000,-1)
xtestFlattened = xtest.reshape(10000,-1)
```

```python
xtrainFlattened = scaler.fit_transform(xtrainFlattened)
xtestFlattened  =scaler.transform(xtestFlattened)
```

```python
xtrain = tf.Variable(xtrain)
ytrain = tf.Variable(ytrain)
xtest = tf.Variable(xtest)
ytest = tf.Variable(ytest)
xtrainFlattened = tf.Variable(xtrainFlattened)
xtestFlattened = tf.Variable(xtestFlattened)
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation = "sigmoid")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(xtrainFlattened,ytrain,epochs=5)
```

```python
ypred = model.predict(xtestFlattened)
```

```python
ypredLabels = [np.argmax(i) for i in ypred]
```

```python
tf.math.confusion_matrix(labels=ytest,predictions=ypredLabels)
```

```python
callback = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=1,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation = "relu",kernel_initializer="he_uniform"),
    Dropout(0.2),
    BatchNormalization(),
    keras.layers.Dense(100,activation = "relu",kernel_initializer="he_uniform"),
    Dropout(0.2),
    BatchNormalization(),
    keras.layers.Dense(10,activation = "sigmoid")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

model.fit(xtrainFlattened,ytrain,epochs=10,callbacks=[callback],batch_size=100,validation_data=(xtestFlattened,ytest))
```

```python
ypred = model.predict(xtestFlattened)
ypredLabels = [np.argmax(i) for i in ypred]
tf.math.confusion_matrix(labels=ytest,predictions=ypredLabels)
```

```python
model.evaluate(xtestFlattened,ytest)
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
    model = keras.Sequential()

    counter = 0

    for i in range(hp.Int("num_layers",min_value=1,max_value=10)):
        if(not counter):
            model.add(
                keras.layers.Dense(          
                    hp.Int(f"units {i}",min_value=8,max_value=128,step=8),
                    activation = hp.Choice(f"activation {i}",values=["relu","tanh","linear","sigmoid"]),
                    input_dim=784,
                    kernel_initializer=hp.Choice(f"kernel {i}",values=["he_normal","he_uniform","glorot_normal","glorot_uniform"])
                )
            )
            model.add(
                Dropout(
                    hp.Choice(f"Dropout {i}",values=[0.2,0.3,0.5,0.6,0.8])
                )
            )
            model.add(BatchNormalization())
            counter+=1
        else:
            model.add(
                keras.layers.Dense(          
                    hp.Int(f"units {i}",min_value=8,max_value=128,step=8),
                    activation = hp.Choice(f"activation {i}",values=["relu","tanh","linear","sigmoid"]),
                    kernel_initializer=hp.Choice(f"kernel {i}",values=["he_normal","he_uniform","glorot_normal","glorot_uniform"])
                )
            )
            model.add(
                Dropout(
                    hp.Choice(f"Dropout {i}",values=[0.2,0.3,0.5,0.6,0.8])
                )
            )
            model.add(BatchNormalization())

    model.add(keras.layers.Dense(10,activation="relu"))
    model.compile(
        optimizer = hp.Choice("optmizer",values=["rmsprop","adadelta","adam","sgd","nadam"]),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model

    

    
```

```python
tuner = kt.GridSearch(build_model,objective="val_accuracy",max_trials=5)
```

```python
tuner.search(xtrainFlattened,ytrain,epochs=10,validation_data=(xtestFlattened,ytest),batch_size=100)
```
